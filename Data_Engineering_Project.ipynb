```python
print("checking missing values:\n")
print(df.isnull().sum())
```

```python
print("Handle missing values:\n")
df = df.dropna()
```

```python
print("Standardize station names:\n")
df['Source_Station_Name'] = df['Source_Station_Name'].str.upper()
df['Destination_Station_Name'] = df['Destination_Station_Name'].str.upper()
```

```python
print("Remove extra spaces:\n")
df['Source_Station_Name'] = df['Source_Station_Name'].str.strip()
df['Destination_Station_Name'] = df['Destination_Station_Name'].str.strip()
```

```python
print(df.columns)
```

```python
print("Task - 2.3: Data Enrichment")
```

```python
print("creating train category column:")
def category(days):
  days = str(days).lower()
  if 'saturday' in days or 'sunday' in days:
    return 'Weekend'
  return 'Weekday'
df['category'] = df['days'].apply(category)
print(df[['Train_No', 'category']].head())
```

```python
print("Heatmap")
import seaborn as sns

top_sources = df['Source_Station_Name'].value_counts().head(10).index
top_destinations = df['Destination_Station_Name'].value_counts().head(10).index

filtered_df = df[
    (df['Source_Station_Name'].isin(top_sources)) &
    (df['Destination_Station_Name'].isin(top_destinations))
]

heatmap_data = pd.crosstab(
    filtered_df['Source_Station_Name'],
    filtered_df['Destination_Station_Name']
)

plt.figure(figsize=(12,8))

sns.heatmap(heatmap_data, annot=True, fmt='d')

plt.title('Source vs Destination Heatmap')

plt.show()
```
